# Example 2 - House Price Prediction (Regression)

# Synthetic dataset

```python
# Example 2 - House Price Prediction (Regression) with Synthetic dataset
# Prdict house prices based on 10 numerical features

import tensorflow as tf
from tensorflow.keras import layers, models
import numpy as np

# Generate synthetic dataset (1000 samples, 10 feature)
np.random.seed(42) # It initializes the random number generator so that random operations produce the same sequence of numbers every time, making results reproducible.
x_train = np.random.rand(1000, 10) # It generates random numbers uniformly distributed between 0 and 1.
                                  # 1000*10 => 10,000 random floating point numbers between 0 and 1
y_train = np.random.rand(1000) * 500000  # House prices between 0-500K


x_test = np.random.rand(200, 10)
y_test = np.random.rand(200) * 500000

# Build model
model = models.Sequential([
    layers.Dense(64, activation='relu', input_shape=(10,)),
    layers.Dense(32, activation='relu'),
    layers.Dense(1) #Single output for regression
])

model.summary()

# Compile model
model.compile(optimizer='adam', loss='mse', metrics=['mae'])

# Train model
model.fit(x_train, y_train, epochs=20, batch_size=32, verbose=1)

#Evaluate model
test_loss, test_mae = model.evaluate(x_test, y_test)
print(f'Test MAE: ${test_mae:.2f}')

#Make a prediction
sample_input = np.random.rand(1, 10)
predicted_price = model.predict(sample_input)
print(f'Predicted House price: ${predicted_price[0][0]:.2f}')
```

# Dataset from scikit-learn

```python
# Example 2 - House Price Prediction (regression) with Dataset

import tensorflow as tf
from tensorflow.keras import layers,models
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import numpy as np

# Load California housing dataset
data = fetch_california_housing()
# Check this link -> https://scikit-learn.org/stable/modules/generated/sklearn.datasets.fetch_california_housing.html to see the actual size of data
# input => data: ndarray, shape (20640, 8) Where each row corresponds to the 8 feature value in order.
# output => numpy array of shape(20640,) value corresponds to the median house value in units of 100,000.

# return_X_y : bool, default=False, if true, return (data.data , data.target) instead of a bunch objects
x,y = data.data, data.target #Feature and target (median house value in 100,000s)

print(f"Target value Range (in 100,0000s): Min = {y.min():.2f}, Max= {y.max():.2f}, Mean ={y.mean():.2f}")

# Split data into train and test sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Normalize features for better training stability Z-score Standarization / Standard scaling
scalar = StandardScaler()
print(f'Scalar : ${scalar}')

x_train = scalar.fit_transform(x_train) #Compute mean & std from training data, then scale
print(f'x_train : ${x_train}')

x_test = scalar.transform(x_test) # Use the same scaling parameters to transform test data

# we use the same mean and standard deviation (computed from x_train) to scale x_test
# This ensure that both training and test data follow the same distribution

# If we include x_test when computing mean and standard deviation, the model gets information from the test set before training
# This is data leakage, which can lead to unrealistically good performance and poor generalization to new data
# Test data should only be used for evaluation after the model is trained.

# In real world applications, new data arrives after testing. we don not get to recompute mean and standard deviation for each new data point
# The model should be able to handle unseen data using the same scaling applied during training

# Build model
model = models.Sequential([
    layers.Dense(64, activation='relu', input_shape = (x_train.shape[1],)), #input features from dataset
    layers.Dense(32, activation='relu'),
    layers.Dense(1) #single output for regression
])

# Compile model
model.compile(optimizer='adam', loss='mse', metrics=['mae'])

# Train model
model.fit(x_train, y_train, epochs=20, batch_size=32, verbose=1)

# Evaluate model
test_loss, test_mae = model.evaluate(x_test, y_test)
print(f'Test MAE: ${test_mae * 100000:.2f}') #Convert to actual dollars

# Make a prediction
sample_input = np.expand_dims(x_test[0], axis=0)  # Take one test sample
predicted_price = model.predict(sample_input)
print(f'Predicted House Price: ${predicted_price[0][0] * 1000000:.2f}') #  Convert to actual dollars

```
