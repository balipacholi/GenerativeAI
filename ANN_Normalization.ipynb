```python
# Normalization Example
# Why Normalization or Scale Down is needed ?
# 1) The activation functions are designed as such its values liew between some sort interval.
# 2) if the difference between the input values is more we need more number of epochs to reach the desired value.
import numpy as np

# Sample data
data = np.array([10, 20, 30, 40, 50, 1000]) # 1000 ia an outlier

# Min-Max Scaling: Scales values between 0 and 1
# When outlier such as 1000 is included in the above data we shouldn't use min max scaling as the results will be incorrect
def min_max_scaling(data):
  min_val = np.min(data) # Min possible value of the input
  max_val = np.max(data) # Max possible value of the input
  return (data - min_val) / (max_val - min_val) # Formula for min max scaling to scale down the input to the range between 0 & 1

def z_score_normalization(data):
  mean = np.mean(data)
  std_dev = np.std(data)
  return (data - mean) / std_dev # Formula for Z score normalization this is used when we have outliers included in our input data

#Apply normalization techniques
min_max_scaled = min_max_scaling(data)
z_score_scaled = z_score_normalization(data)

# Print results
print("Original Data:",data)
# For min max scaling output is  for the above input data is : [0.         0.01010101 0.02020202 0.03030303 0.04040404 1.        ]
# Observe all the other min max scaling value is around 0. but for outlier it is 1.
print("min max scaling data:",min_max_scaled)
print("z score standarized data:",z_score_scaled)
```
